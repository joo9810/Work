```python
import json

def print_input_values(id, main_node_id, first_node_id, second_node_id, third_node_id, node_flow,
                       ipzy1, ipzy2, ipzy3, name1, name2, name3, name4, transactions_data):
    # # 정해야 하는 값
    # id = 15
    # main_node_id   = "1000000000001"
    # first_node_id  = "2000000000001"
    # second_node_id = "4000000000001"
    # third_node_id  = "5000000000001"
    # node_flow = "1-1-1" # ex) 2-1-2, 1-2

    # 알아서 정해지는 값
    graph_name = f"테스트그래프-{id}"
    start_date = '2024-01-01'
    end_date = '2024-03-31'
    flow_parts = node_flow.split("-")

    main_node_acc_id = main_node_id[-1]
    first_node_acc_id = first_node_id.split("000000000001")[0]
    second_node_acc_id = second_node_id.split("000000000001")[0]
    thrid_node_acc_id = third_node_id.split("000000000001")[0]

    a_dict = {
        "account" : {
            "id" : f"main{main_node_acc_id}",
            "properties" : {
                "acno" : main_node_id,
                "name" : f"{name1}",
                "bnkcd" : "004"
            }
        }
    }
    a = f"'{json.dumps(a_dict, ensure_ascii=False)}'"
    #--------------------------------------------------------
    if len(flow_parts) >= 1:
        b_dict = {
            "account" : {
                "id" : f"acc{first_node_acc_id}",
                "properties" : {
                    "acno" : first_node_id,
                    "name" : f"{name2}",
                    "bnkcd" : "004"
                }
            }
        }
        b = f"'{json.dumps(b_dict, ensure_ascii=False)}'"
        # ensure_ascii=False : 한글이 유니코드 이스케이프 안되고 그대로 나옴

        if ipzy1 == 1:
            r1_id = f"main{main_node_acc_id}acc{first_node_acc_id}_1"
        elif ipzy1 == 2:
            r1_id = f"main{main_node_acc_id}acc{first_node_acc_id}"

        r1_dict = {
        "transact" : {
            "id" : r1_id,
            "end" : f"acc{first_node_acc_id}",
            "start" : f"main{main_node_acc_id}",
            "properties" : {
                "ipzygubun" : f"{ipzy1}"
                }
            }
        }
        r1 = f"'{json.dumps(r1_dict, ensure_ascii=False)}'"
        aggr_val1 = f"'{'6,2350617.000000000000000,14103702.000'}'"
    else:
        b = "NULL"
        r1 = "NULL"
        aggr_val1 = "NULL"
    #--------------------------------------------------------
    if len(flow_parts) >= 2:
        c_dict = {
            "account" : {
                "id" : f"acc{second_node_acc_id}",
                "properties" : {
                    "acno" : second_node_id,
                    "name" : f"{name3}",
                    "bnkcd" : "004"
                }
            }
        }
        c = f"'{json.dumps(c_dict, ensure_ascii=False)}'"

        if ipzy2 == 1:
            r2_id = f"main{first_node_acc_id}acc{second_node_acc_id}_1"
        elif ipzy2 == 2:
            r2_id = f"main{first_node_acc_id}acc{second_node_acc_id}"

        r2_dict = {
        "transact" : {
            "id" : r2_id,
            "end" : f"acc{second_node_acc_id}",
            "start" : f"acc{first_node_acc_id}",
            "properties" : {
                "ipzygubun" : f"{ipzy2}"
                }
            }
        }
        r2 = f"'{json.dumps(r2_dict, ensure_ascii=False)}'"
        aggr_val2 = f"'{'6,5559369.000000000000000,33356214.000'}'"
    else:
        c = "NULL"
        r2 = "NULL"
        aggr_val2 = "NULL"
    #--------------------------------------------------------
    if len(flow_parts) == 3:
        d_dict = {
            "account" : {
                "id" : f"acc{thrid_node_acc_id}",
                "properties" : {
                    "acno" : third_node_id,
                    "name" : f"{name4}",
                    "bnkcd" : "004"
                }
            }
        }
        d = f"'{json.dumps(d_dict, ensure_ascii=False)}'"

        if ipzy3 == 1:
            r3_id = f"trans{id}_1"
        elif ipzy3 == 2:
            r3_id = f"trans{id}"

        r3_dict = {
        "transact" : {
            "id" : r3_id,
            "end" : f"acc{thrid_node_acc_id}",
            "start" : f"acc{second_node_acc_id}",
            "properties" : {
                "ipzygubun" : f"{ipzy3}"
                }
            }
        }
        r3 = f"'{json.dumps(r3_dict, ensure_ascii=False)}'"
        aggr_val3 = f"'{transactions_data}'"
    else:
        d = "NULL"
        r3 = "NULL"
        aggr_val3 = "NULL"

    created_at = '2024-04-24 15:43:44.600'
    updated_at = '2024-04-24 15:43:44.600'

    print(f"{id}, '{graph_name}', '{main_node_id}', '{start_date}', '{end_date}',\n{a},\n{b},\n{r1}, {aggr_val1},\n{c},\n{r2}, {aggr_val2},\n'{created_at}', '{updated_at}',\n{d},\n{r3}, {aggr_val3}")
```
